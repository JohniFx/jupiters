```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('https://www.ksh.hu/stadat_files/nep/hu/nep0065.xlsx', skiprows=1)
df.tail(3)
```

```python
df.index = df.set_index('A hét kezdő napja')
```

```python
df.index=pd.date_range(start=df['A hét kezdő napja'][0], freq='W', periods=422)
```

```python
df.columns
```

```python
for c in df.columns[4:]:
    try:
        df[c].plot(title=c, figsize=(20,4))
        plt.show()
    except Exception as e:
        print(c, e)   
```
